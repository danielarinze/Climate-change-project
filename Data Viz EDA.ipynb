```python
# Assuming `Temperature`, `Crop Yield`, `Region`, and `Crop Type` columns exist in your dataset.

# Group by Region and Crop Type, then calculate the averages
df1 = (
    sample_data.groupby(['Continents', 'Crop_Category'])
    .agg(Average_Temperature=('Average_Temperature_C', 'mean'), 
         Average_Crop_Yield=('Crop_Yield_MT_per_HA', 'mean'))
    .reset_index()
)

# Display the resulting dataframe
df1
```

```python
correlation = sample_data['Average_Temperature_C'].corr(sample_data['Crop_Yield_MT_per_HA'])
print(f"Correlation between Temperature and Crop Yield: {correlation}")
```

```python
# Group by Continents and Crop_Category, then calculate totals and averages
df2 = (
    sample_data.groupby(['Continents', 'Crop_Category'])
    .agg(Total_Precipitation_mm=('Total_Precipitation_mm', 'sum'), 
         Average_Crop_Yield_MT_per_HA=('Crop_Yield_MT_per_HA', 'mean'))
    .reset_index()
)
# Display the dataframe
df2
```

```python
correlation_precip = sample_data['Total_Precipitation_mm'].corr(sample_data['Crop_Yield_MT_per_HA'])
print(f"Correlation between Precipitation and Crop Yield: {correlation_precip}")
```

```python
# Group by Continents and Crop_Category to calculate aggregated values
df3 = (
    sample_data.groupby(['Continents', 'Crop_Category'])
    .agg(
        Total_CO2_Emissions_MT=('CO2_Emissions_MT', 'sum'),
        Total_Economic_Impact_Million_USD=('Economic_Impact_Million_USD', 'sum'),
        Average_Crop_Yield_MT_per_HA=('Crop_Yield_MT_per_HA', 'mean')
    )
    .reset_index()
)



# Display the datafr
df3
```

```python
import matplotlib.pyplot as plt

# Scatter plot with bubble size
plt.figure(figsize=(10, 6))
plt.scatter(
    df3['Total_CO2_Emissions_MT'], 
    df3['Total_Economic_Impact_Million_USD'], 
    s=df3['Average_Crop_Yield_MT_per_HA'] * 50,  # Scale bubble size
    alpha=0.6,
    edgecolors='w'
)

# Adding labels and title
plt.xlabel('Total CO2 Emissions (MT)')
plt.ylabel('Total Economic Impact (Million USD)')
plt.title('CO2 Emissions vs Economic Impact (Bubble Size: Crop Yield)')
plt.grid(True)
plt.show()
```

```python
# Count frequency of extreme weather events and unique adaptation strategies by Continents and Crop_Category
df4 = (
    sample_data.groupby(['Continents', 'Crop_Category'])
    .agg(
        Frequency_Extreme_Weather_Events=('Extreme_Weather_Events', 'sum'),
        Unique_Adaptation_Strategies=('Adaptation_Strategies', lambda x: len(set(x)))
    )
    .reset_index()
)
# Display the dataframe
df4
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Pivot the data for heatmap
heatmap_data = df4.pivot('Continents', 'Crop_Category', 'Frequency_Extreme_Weather_Events')

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(
    heatmap_data, 
    annot=True, 
    fmt="d", 
    cmap="YlGnBu", 
    cbar_kws={'label': 'Frequency of Extreme Weather Events'}
)

# Add title and labels
plt.title('Heat Map: Extreme Weather Events vs Adaptation Strategies')
plt.xlabel('Crop Category')
plt.ylabel('Continents')
plt.show()
```

```python
# Group data by Year and calculate averages
df5 = (
    sample_data.groupby('Year')
    .agg(
        Average_Irrigation_Access_Percentage=('Irrigation_Access_%', 'mean'),
        Average_Crop_Yield_MT_per_HA=('Crop_Yield_MT_per_HA', 'mean')
    )
    .reset_index()
)


# Display the dataframe
df5
```

```python
import matplotlib.pyplot as plt

# Plot the line graph
plt.figure(figsize=(12, 6))

# Line for Irrigation Access
plt.plot(
    df5['Year'], 
    df5['Average_Irrigation_Access_Percentage'], 
    label='Irrigation Access (%)', 
    color='blue', 
    marker='o'
)

# Line for Crop Yield
plt.plot(
    df5['Year'], 
    df5['Average_Crop_Yield_MT_per_HA'], 
    label='Crop Yield (MT/HA)', 
    color='green', 
    marker='o'
)

# Adding title and labels
plt.title('Irrigation Access vs Crop Yield Over the Years')
plt.xlabel('Year')
plt.ylabel('Percentage / Yield')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Group data by Region and Crop Type to calculate averages
df6 = (
    sample_data.groupby(['Continents', 'Crop_Category'])
    .agg(
        Average_Pesticide_Use_KG_per_HA=('Pesticide_Use_KG_per_HA', 'mean'),
        Average_Fertilizer_Use_KG_per_HA=('Fertilizer_Use_KG_per_HA', 'mean'),
        Average_Soil_Health_Index=('Soil_Health_Index', 'mean')  # Include soil health insight
    )
    .reset_index()
)

# Save the resulting dataframe
df6.to_csv("Pesticide_and_Fertilizer_Use.csv", index=False)

# Display the dataframe
print(df6)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Pivot data to reshape for bar chart
pesticide_data = df6.pivot(index='Continents', columns='Crop_Category', values='Average_Pesticide_Use_KG_per_HA')
fertilizer_data = df6.pivot(index='Continents', columns='Crop_Category', values='Average_Fertilizer_Use_KG_per_HA')

# Prepare x positions for bars
regions = df6['Continents'].unique()
x = np.arange(len(regions))

# Set bar width
bar_width = 0.6

# Initialize the figure
fig, ax = plt.subplots(figsize=(14, 8))

# Plot stacked bars for each crop category
bottom_pesticide = np.zeros(len(x))  # Start stacking from 0
for crop in pesticide_data.columns:
    ax.bar(
        x, 
        pesticide_data[crop], 
        width=bar_width, 
        bottom=bottom_pesticide, 
        label=f'Pesticide Use - {crop}', 
        color=np.random.rand(3,)
    )
    bottom_pesticide += pesticide_data[crop]

bottom_fertilizer = np.zeros(len(x))  # Start stacking from 0
for crop in fertilizer_data.columns:
    ax.bar(
        x, 
        fertilizer_data[crop], 
        width=bar_width, 
        bottom=bottom_fertilizer, 
        label=f'Fertilizer Use - {crop}', 
        color=np.random.rand(3,)
    )
    bottom_fertilizer += fertilizer_data[crop]

# Set labels and title
ax.set_xticks(x)
ax.set_xticklabels(regions)
ax.set_xlabel('Regions')
ax.set_ylabel('Usage (KG/HA)')
ax.set_title('Pesticide and Fertilizer Use by Region and Crop Type')
ax.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Group data by Region to calculate the average Soil Health Index
df7 = (
    sample_data.groupby('Continents')
    .agg(
        Average_Soil_Health_Index=('Soil_Health_Index', 'mean')
    )
    .reset_index()
)

# Save the resulting dataframe
df7.to_csv("Soil_Health_Index_by_Region.csv", index=False)

# Display the dataframe
print(df7)

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Number of variables (regions in this case)
categories = df7['Continents'].tolist()
values = df7['Average_Soil_Health_Index'].tolist()

# Create the radar chart
num_vars = len(categories)

# Split the circle into even parts and save the angles for each category
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# The radar chart needs to be closed, so we append the first value at the end
values += values[:1]
angles += angles[:1]

# Plotting
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

# Create the radar chart
ax.fill(angles, values, color='green', alpha=0.25)
ax.plot(angles, values, color='green', linewidth=2)  # Outline the area

# Add labels
ax.set_yticklabels([])  # Remove radial ticks for clarity
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)

# Add a title
ax.set_title('Soil Health Index Across Regions')

plt.tight_layout()
plt.show()
```
